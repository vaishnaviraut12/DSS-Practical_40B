```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
 x=np.arange(1,11)
```

```python
x
```

```python
 y=2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
